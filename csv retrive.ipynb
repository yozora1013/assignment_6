```python
fo = open("liveshow_timetable.csv", "r")
ls = []
for line in fo:
    line = line.replace("\n", "")
    ls = line.split(",")
    
    lns = ""
    for s in ls:
        lns += "{}\t".format(s)
    print(lns)

fo.close()


seg1 = '''<!DOCTYPE html><html lang="en"><head>
<meta charset="GBK">
<title>Title</title>
</head>
<body>

<table border="1" align="center" width="100%">\n
<tr bgcolor="grey">\n'''


seg2 = "</tr>\n"
seg3 = "</table>\n</body>\n</html>"


def fill_data(locls):
    seg = '<tr><td align="center">{}</td><td align="center">{}</td><td align="center">{}</td><td align="center">{}</td><td align="center">{}</td><td align="center">{}</td><td align="center">{}</td><td align="center">{}</td></tr>\n'.format(*locls)
    return seg


ls = []
with open("liveshow_timetable.csv", "r") as fr:
    for line in fr:
        line = line.replace("\n", "")  
        ls.append(line.split(","))  


with open("liveshow_timetable.html", "w") as fw:
    fw.write(seg1)
    
    fw.write('<th width="15%">{}</th>\n<th width="15%">{}</th>\n<th width="15%">{}</th>\n<th width="15%">{}</th>\n<th width="15%">{}</th>\n<th width="15%">{}</th>\n<th width="15%">{}</th>\n<th width="15%">{}</th>\n'.format(*ls[0]))
    fw.write(seg2)

    for i in range(1, len(ls)):
        fw.write(fill_data(ls[i]))
 
    fw.write(seg3)
```
